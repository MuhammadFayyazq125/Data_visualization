# Task10

Task 10: Calculate the covariance and correlation coefficient between Price and Area, no matter if the apartment is sold or not. Is the result in line with the scatter plot?	
	
	
Solution:	
	
Covariance	24147721.73
Correlation	0.951087377
	
Yes, the result is in line with the scatter plot. The two variables are greatly correlated. 	


```python
covariance = data['Price'].cov(data['Area (ft.)'])
correlation = data['Price'].corr(data['Area (ft.)'])
plt.scatter(covariance,correlation, alpha = 1)
plt.show()
print(f'covariance = {covariance} \n correlation = {correlation}')
```
